```python
# Cargar todas las bibliotecas necesarias

from fastai.vision import *
from fastai.metrics import error_rate, accuracy

import warnings
warnings.filterwarnings('ignore')

from google.colab import drive
drive.mount('/content/drive')
```

```python
# Establecemos la ruta a las imagenes de malware y no-malware
# Blanco y Negro
path = Path('/content/drive/My Drive/Clasificador_Malware/Muestras_ByN')
path.ls()
```

```python
#Todas las arquitecturas imágenes 64x64
data = ImageDataBunch.from_folder(path, train='train', valid='test', ds_tfms=get_transforms(do_flip=False), size=(64,64), bs=64, num_workers=8)

#ResNet18
learn = cnn_learner(data, models.resnet18, metrics = [accuracy,error_rate,Precision(),Recall(),FBeta()])
learn.fit_one_cycle(25)
learn.recorder.plot_metrics()
interp = ClassificationInterpretation.from_learner(learn)
interp.plot_confusion_matrix(figsize=(6,6), dpi=60)
learn.unfreeze()
learn.lr_find()
learn.recorder.plot(suggestion=True)
#Reemplazar XX por el valor adecuado
learn.fit_one_cycle (5, max_lr = slice (XX, XX))

#ResNet34
learn = cnn_learner(data, models.resnet34, metrics = [accuracy,error_rate,Precision(),Recall(),FBeta()])
learn.fit_one_cycle(25)
learn.recorder.plot_metrics()
interp = ClassificationInterpretation.from_learner(learn)
interp.plot_confusion_matrix(figsize=(6,6), dpi=60)
learn.unfreeze()
learn.lr_find()
learn.recorder.plot(suggestion=True)
#Reemplazar XX por el valor adecuado
learn.fit_one_cycle (5, max_lr = slice (XX, XX))

#ResNet50
learn = cnn_learner(data, models.resnet50, metrics = [accuracy,error_rate,Precision(),Recall(),FBeta()])
learn.fit_one_cycle(25)
learn.recorder.plot_metrics()
interp = ClassificationInterpretation.from_learner(learn)
interp.plot_confusion_matrix(figsize=(6,6), dpi=60)
learn.unfreeze()
learn.lr_find()
learn.recorder.plot(suggestion=True)
#Reemplazar XX por el valor adecuado
learn.fit_one_cycle (5, max_lr = slice (XX, XX))

#ResNet152
learn = cnn_learner(data, models.resnet152, metrics = [accuracy,error_rate,Precision(),Recall(),FBeta()])
learn.fit_one_cycle(25)
learn.recorder.plot_metrics()
interp = ClassificationInterpretation.from_learner(learn)
interp.plot_confusion_matrix(figsize=(6,6), dpi=60)
learn.unfreeze()
learn.lr_find()
learn.recorder.plot(suggestion=True)
#Reemplazar XX por el valor adecuado
learn.fit_one_cycle (5, max_lr = slice (XX, XX))

#VGG16
learn = cnn_learner(data, models.vgg16_bn, metrics = [accuracy,error_rate,Precision(),Recall(),FBeta()])
learn.fit_one_cycle(25)
learn.recorder.plot_metrics()
interp = ClassificationInterpretation.from_learner(learn)
interp.plot_confusion_matrix(figsize=(6,6), dpi=60)
learn.unfreeze()
learn.lr_find()
learn.recorder.plot(suggestion=True)
#Reemplazar XX por el valor adecuado
learn.fit_one_cycle (5, max_lr = slice (XX, XX))

#DenseNet121
learn = cnn_learner(data, models.densenet121, metrics = [accuracy,error_rate,Precision(),Recall(),FBeta()])
learn.fit_one_cycle(25)
learn.recorder.plot_metrics()
interp = ClassificationInterpretation.from_learner(learn)
interp.plot_confusion_matrix(figsize=(6,6), dpi=60)
learn.unfreeze()
learn.lr_find()
learn.recorder.plot(suggestion=True)
#Reemplazar XX por el valor adecuado
learn.fit_one_cycle (5, max_lr = slice (XX, XX))

#DenseNet201
learn = cnn_learner(data, models.densenet201, metrics = [accuracy,error_rate,Precision(),Recall(),FBeta()])
learn.fit_one_cycle(25)
learn.recorder.plot_metrics()
interp = ClassificationInterpretation.from_learner(learn)
interp.plot_confusion_matrix(figsize=(6,6), dpi=60)
learn.unfreeze()
learn.lr_find()
learn.recorder.plot(suggestion=True)
#Reemplazar XX por el valor adecuado
learn.fit_one_cycle (5, max_lr = slice (XX, XX))

#AlexNet
learn = cnn_learner(data, models.alexnet, metrics = [accuracy,error_rate,Precision(),Recall(),FBeta()])
learn.fit_one_cycle(25)
learn.recorder.plot_metrics()
interp = ClassificationInterpretation.from_learner(learn)
interp.plot_confusion_matrix(figsize=(6,6), dpi=60)
learn.unfreeze()
learn.lr_find()
learn.recorder.plot(suggestion=True)
#Reemplazar XX por el valor adecuado
learn.fit_one_cycle (5, max_lr = slice (XX, XX))
```
